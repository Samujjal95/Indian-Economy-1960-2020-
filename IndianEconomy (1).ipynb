```python
from statsmodels.tsa.arima.model import ARIMA
import warnings
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error
from math import sqrt
warnings.filterwarnings('ignore')
```

```python
# Check for stationarity using ADF test
result = adfuller(df['GDP growth (annual %)'])
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
```

```python
# Function to create and evaluate ARIMA model
def arima_forecast(series, order,steps=5):
    train = series[:int(0.8 * len(series))]
    test = series[int(0.8 * len(series)):]
    model = ARIMA(train, order=order)
    results = model.fit()
    forecast = results.forecast(steps=len(test)+steps)
    rmse = sqrt(mean_squared_error(test, forecast[:len(test)]))
    return forecast, rmse
```

```python
# Forecast GDP growth
gdp_forecast, gdp_rmse = arima_forecast(df['GDP growth (annual %)'], order=(1,1,1),steps=5)

# Forecast inflation
inflation_forecast, inflation_rmse = arima_forecast(df['Inflation, consumer prices (annual %)'], order=(1,1,1),steps=5)

future_years = pd.date_range(start=df.index[-1], periods=6, freq='Y')[1:]

# Plot forecasts
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['GDP growth (annual %)'], label='Actual GDP Growth')
plt.plot(df.index[int(0.8 * len(df)):], gdp_forecast[:len(df) - int(0.8 * len(df))], label='Forecasted GDP Growth')
plt.plot(future_years, gdp_forecast[-5:], marker='o', color='r', label='5-Year GDP Growth Forecast')
plt.title('GDP Growth Forecast')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Inflation, consumer prices (annual %)'], label='Actual Inflation')
plt.plot(df.index[int(0.8 * len(df)):], inflation_forecast[:len(df) - int(0.8 * len(df))], label='Forecasted Inflation')
plt.plot(future_years, inflation_forecast[-5:], marker='o', color='r', label='5-Year Inflation Forecast')
plt.title('Inflation Forecast')
plt.legend()
plt.show()
```
